```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt

df = pd.read_csv('./data/game.csv')

wl_home = 'wl_home'
game_date = 'game_date'
team_abv = 'team_abbreviation_home'

new_df = df[[wl_home, game_date, team_abv ]]
pd.to_datetime(new_df[game_date])
new_df['month'] = pd.to_datetime(new_df[game_date]).dt.month_name()
df_years = pd.DatetimeIndex(new_df[game_date])

team = 'CHI'

bulls_df = new_df.loc[(new_df[team_abv] == team) & (new_df[game_date].between("2022-01-01", "2022-12-31"))]
bulls_df[wl_home] = bulls_df[wl_home].map({"W": 1.0, "L": 0.0})

bulls_df.plot(kind='line', x ='game_date', y ='wl_home', figsize=(15, 4))

plt.legend([2022], loc = 'lower right', title = 'Year')
plt.ylabel('W(1)-L(0)')
plt.xlabel('Year_Month_date')
plt.title('Chicago Bulls games 2022')
```
